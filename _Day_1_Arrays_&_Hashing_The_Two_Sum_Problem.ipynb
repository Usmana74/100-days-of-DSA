# **1. The Array (The Foundation)**
An Array (or list in Python) is a collection of items stored in contiguous memory locations.

**Strengths:** Accessing an element by index is incredibly fast ($O(1)$).

**Weaknesses:** Searching for a value (if you don't know the index) is slow ($O(n)$) because you have to check every "bucket" one by one.2.

# **The Hash Map (The Speed Booster)**
A Hash Map (or dict in Python) uses a Hash Function to map keys to specific values.

**The difference:** It allows us to find a value instantly. Instead of walking through a list, the Hash Map tells us exactly where the data is.

**Big O:** Search, Insert, and Delete are all $O(1)$ on average.

# **3. The "Trade-off" Principle**

In DSA, we often face a choice:

**Brute Force:** Use little memory but take a lot of time (e.g., $O(n^2)$).

**Optimized: **Use extra memory (a Hash Map) to make the code run significantly faster (e.g., $O(n)$).

**This is the core of most Array-based LeetCode problems.**

# **Two Sum (LeetCode #1)**
 https://leetcode.com/problems/two-sum/

#  ðŸ“ Problem Statement

Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to `target`.
- Each input has exactly one solution.
- You may not use the same element twice.

### ðŸ’¡ Intuition
The "Brute Force" way is to use two nested loops to check every pair, but that takes $O(n^2)$ time. To optimize, we use a **Hash Map (Dictionary)** to store the numbers we've already seen. As we move through the list, we calculate the  ($target - current\_number$). If that complement is in our map, we have our answer!

# **ðŸ” Dry Run**
**Input:** `nums = [2, 7, 11, 15]`, `target = 9`

1. **i = 0 (val: 2):** Need `9 - 2 = 7`. Is 7 in `prevMap`? No.
   - Add 2 to map: `prevMap = {2: 0}`
2. **i = 1 (val: 7):** Need `9 - 7 = 2`. Is 2 in `prevMap`? **YES!**
   - Return indices: `[prevMap[2], 1]` -> `[0, 1]`

# **Code Solution**

```python
class Solution(object):

    def twoSum(self, nums, target):

        # Initialize an empty dictionary to store numbers we have already seen.
        # Format: { value: index }
        prevHashM = {}

        # Iterate through the list using enumerate to get both index (i) and value (n)
        for i, n in enumerate(nums):

            # Calculate the required number to reach the target
            diff = target - n

            # Check if the required number is already in our dictionary
            if diff in prevHashM:

                # If found, return the index of the 'diff' and the current index 'i'
                return [prevHashM[diff], i]

            # If not found, add the current number to the dictionary
            # We use the number as the KEY and its index as the VALUE
            prevHashM[n] = i

# --- Local Testing ---
if __name__ == "__main__":
    solver = Solution()

    # Test Case 1
    nums = [2, 7, 11, 15]
    target = 9
    result = solver.twoSum(nums, target)
    print(f"Test Case 1: nums={nums}, target={target} -> Result: {result}")

    # Test Case 2
    nums2 = [3, 2, 4]
    target2 = 6
    result2 = solver.twoSum(nums2, target2)
    print(f"Test Case 2: nums={nums2}, target={target2} -> Result: {result2}")
```

## **--- Complexity Analysis ---**
 Time Complexity: O(n) - We only traverse the list once.

 Space Complexity: O(n) - In the worst case, we store n elements in the hash map.
